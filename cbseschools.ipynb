```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

#later
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
```

```python
options = Options()
#options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
#driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.get('https://saras.cbse.gov.in/cbse_aff/schdir_Report/userview.aspx')
time.sleep(3)
```

```python
affiliation_level_dropdown = driver.find_element(By.ID, 'optlist_4')
affiliation_level_dropdown.click()
school_level = Select(driver.find_element(By.ID, 'ddlitem'))  
school_level.select_by_visible_text('Senior Secondary Level')
#keyword 'a' in the search input box
time.sleep(3)
submit_button = driver.find_element(By.ID, 'search')
submit_button.click()

search_input = driver.find_element(By.ID, 'keytext')
search_input.send_keys('a')
submit_button = driver.find_element(By.ID, 'search')
submit_button.click()
time.sleep(3)
```

```python
a=0
dflist = []
pn=0
while a==0:
    print(f'page {pn+1}')
    pn+=1
    time.sleep(3)
    table = driver.page_source
    soup = BeautifulSoup(table, 'html.parser')
    table = soup.find('table', {'id': 'T1'})
    df=pd.read_html(str(table))
    df = df[0]
    dfo = pd.DataFrame()
    name = []
    email = []
    rowcounter = 4
    for i in range(df.shape[0]):
        if rowcounter>227:
            break
        name.append(df.iloc[rowcounter,0])
        email.append(df.iloc[rowcounter+6,0])
        rowcounter+=9
    dfo['name'] = name
    dfo['email'] = email
    dflist.append(dfo)
    try:
        next_button = driver.find_elements(By.CLASS_NAME, 'buttons')
        for n in next_button:
            if n.get_attribute('value')=='Next >>':
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                n.click()
            else:
                print(n.get_attribute('value'))
    except Exception as e:
        print('exiting')
        print(e)
        a=1
```

```python
dfo= pd.concat(dflist)
```

```python
dfo['name'] = dfo['name'].apply(lambda x: x.split('Name: ')[1])
```

```python
def find_mail(s):
    a = re.findall('\S+@\S+', s)
    if len(a)>0:
        return a[0]
    else:
        return pd.NA
```

```python
import re
dfo['email'] = dfo['email'].apply(find_mail)
```
